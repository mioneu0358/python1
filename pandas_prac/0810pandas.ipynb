```python
df = pd.read_csv("Koweps_hpc10_2015_beta1.csv")
codebook = pd.read_excel("Koweps_Codebook.xlsx")
```

```python
nc = dict(codebook[["변수명","설명"]].values)
df.rename(columns=nc,inplace=True)
val = list(nc.values())
df[val]
```

```python
# 성별 전처리
# 1 = 남  2 = 여  9 = Nan

# temp = {1:"남",2:"여"}

temp = {}
for i in codebook["내용"][0].split():
    a = i.split('.')
    temp[int(a[0])] = a[1]
    
df["성별"] = df.apply(lambda x: temp[x.성별] if x.성별 <= 2 else pd.NA, axis=1)
```

```python
# 0 = 비해당(18세 미만)  1 = 유배우  2 = 사별  3 = 이혼  4 = 별거  5 = 미혼(18세이상, 미혼모 포함)  6 = 기타(사망 등)

# temp2 = {0:"비해당(18세 미만)",
#          1:"유배우",
#          2:"사별",
#          3:"이혼",
#          4:"별거",
#          5:"미혼(18세 이상, 미혼모 포함)",
#          6:"기타(사망 등)"}

temp = []
temp2 = {}
for i in codebook["내용"][2].split():
    temp.extend(i.split("("))
for i in temp:
    if "." in i:
        t = i.split(".")
        temp2[int(t[0])] = t[1]

df["혼인상태"] = df.apply(lambda x: temp2[x.혼인상태] if x.혼인상태 <= 6 else pd.NA, axis=1)
df[val]


# import re
# marriage_data = codebook['내용'][2]
# pattern = r'((\d+)\.(.*?)(?=\s\d+\.|$))' 
# # (\d+)= \d: 숫자로시작, +:1개이상, (): 이 부분을 결과로 반환
# # \.: 마침표찾기,
# # (.*?) = .: 어떤 문자든 1개, *: 1개이상반복, ?: 가능한 짧게 
# # (?=\s\d+\.|$) = (?= ...): lookahead라고 불리며, ...에 들어간 패턴이 뒤 따라오는 경우에만을 의미
# # \s: 공백문자, \d+: 숫자하나이상, \.마침표, |: 또는 $: 문자열 끝
# marriage_data = re.findall(pattern=pattern,string=marriage_data)
# df['혼인상태'] = df["혼인상태"].map({int(k):v.strip() for s,k,v in marriage_data})
# df[val]
```

```python
gmi = df.groupby(["성별","혼인상태"]).size().unstack("혼인상태")
gmi["총 인원"] = gmi.sum(axis=1)
gmi["혼인한 사람"] = gmi["총 인원"] - gmi["비해당"] - gmi["미혼"]
gmi["이혼율"] = (gmi["이혼"] / gmi['혼인한 사람'] * 100).round(2)
gmi
```
