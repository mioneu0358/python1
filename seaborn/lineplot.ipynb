```python
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

sns.set(style = 'darkgrid')
```

<h1> 라인플롯(Line Plot)</h1>

```python
flights = sns.load_dataset("flights")   # flights: 연도별 승객 수치
flights
```

```python
sns.relplot(x='year', y="passengers", kind='line', data=flights)  # kind: plot 종류 설정
```

```python
dots = sns.load_dataset('dots')   
dots
```

```python
sns.relplot(x="time",y="firing_rate", kind='line', data=dots)   
```

```python
# None 입력시 표시 삭제, "sd" 입력시 표준편차 
sns.relplot(x="time",y="firing_rate", kind='line', errorbar=None ,data=dots)                                                               
```

```python
# "sd" 입력시 표준편차 (standard deviation)
sns.relplot(x="time",y="firing_rate", kind='line', errorbar="sd" ,data=dots)
```

```python
# estimator=None: 추정회귀선 비활성화,
sns.relplot(x="time",y="firing_rate", estimator=None, 
            errorbar="sd", kind='line' ,data=dots) 
```

```python
sns.relplot(x="time",y="firing_rate", hue="choice", 
             errorbar="sd" ,
             kind='line',data=dots) 
```

```python
sns.relplot(x="time",y="firing_rate", hue="align",errorbar="sd" ,
            kind='line',data=dots) 
```

```python
sns.relplot(x="time",y="firing_rate", hue="align", 
             errorbar="sd", style='choice',
            kind='line', data=dots) 
```

```python
# dashes = False: 파선(- - -) 해제, markers = True: 기호가 들어간 점선 선택
sns.relplot(x="time",y="firing_rate", hue="align", 
            errorbar="sd", style='choice', 
            dashes=False, markers=True,
            kind='line',data=dots) 
```

```python
# col항목 열어 데이터 분리
sns.relplot(x="time",y="firing_rate", hue="align", 
            errorbar="sd", col='choice', 
            dashes=False, markers=True,
            kind='line', data=dots) 
```

```python
# query() 함수: pandas 에서 데이터프레임을 쿼리하고 원하는 조건에 맞는 데이터를 선택하는 데 사용되는 메서드이다. 
sns.relplot(x="time",y="firing_rate" ,
            style='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
# hue를 사용해서 데이터 세분화
sns.relplot(x="time",y="firing_rate" ,
            hue = "coherence",style='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
# col을 사용해서 데이터 세분화
sns.relplot(x="time",y="firing_rate" ,
            hue = "coherence",col='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = "timepoint", y="signal",
          kind = 'line',data = fmri)
```

```python
# sort=False: 정렬 취소, seaborn이 데이터를 정렬시킨 후 보여준다.
sns.relplot(x = "timepoint", y="signal", sort=False,  # 
          kind = 'line',data = fmri)
```

```python
sns.relplot(x = "timepoint", y="signal",
            style='region',size='event',
          kind = 'line',data = fmri)
```

```python
# 많은 데이터를 같은 화면에 담아두는것은 좋지 않다는 예
sns.relplot(x = "timepoint", y="signal",
           hue="subject", style='region',size='event',
          kind = 'line',data = fmri)
```

```python
# col로 분리
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',size='event',
          kind = 'line',data = fmri)
```

```python
# palette: 색상 정의 후 적용
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',style='event',
           palette = palette, kind = 'line',data = fmri)
```

```python
# # https://velog.io/@imindude/%ED%8C%8C%EC%9D%B4%EC%8D%AC-seaborn-%EB%AA%A8%EB%93%88%EC%9D%98-color-palette
# palette = sns.color_palette('winter', 20) 
# sns.palplot(palette)
```

```python
# cue 데이터만 보기
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',
           palette = palette, kind = 'line',data = fmri.query("event == 'cue'"))
```

```python
# row 추가
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',row = "event",
           palette = palette, kind = 'line',data = fmri)
```

```python
# subject별로 구분하기
sns.relplot(x = "timepoint", y="signal",
           hue="event", style="event",
           col='subject', col_wrap=4,
           kind = 'line', data = fmri.query("region == 'parietal'"))
```

<h3> 데이터프레임으로 정의해보기</h3>

```python
# random.randn() 함수는 표준정규분포 (Standard normal distribution)로부터 샘플링된 난수를 반환
randN = np.random.randn(5,2)   # np.random.rand(x): 1차원 , np.random.rand(x,y): 2차원(x행,y열)
randN
```

```python
tdf = pd.DataFrame(np.random.randn(40,4),#
                   index = pd.date_range('2024-01-01',periods=40),  # date_range('시작연-월-일', periods=일수)
                   columns =['A','B','C','D'])
# print(tdf)
tdf
```

```python
sns.relplot(kind='line',data = tdf)
```

```python
g = sns.relplot(kind='line',data = tdf)
g.fig.autofmt_xdate()   # gigure.autoformat_xdate
```

```python
# 'A'컬럼데이터만 뽑아오겠다. 
A= sns.relplot(kind='line',data = tdf['A'])
A.fig.autofmt_xdate()   # gigure.autoformat_xdate
```
