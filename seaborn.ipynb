```python
import seaborn as sns
import matplotlib as mpl
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

sns.set(style = 'darkgrid')
```

<h1>산점도(Scatter Plot)</h1>

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
```

```python
sns.relplot(x = "flipper_length_mm", y ="body_mass_g", data=penguins )
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm",
            hue = "bill_length_mm", data=penguins)  # hue : 데이터 수치에 따른 색상적용
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm",
            hue = "species", style="species", data=penguins)  # style: 데이터에 구분에 따라 다른 스타일 적용
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm",
            hue = "species", col="island", data=penguins)  # col: 넘긴 기준별로 그려준다.
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm",
            hue = "species", size = "body_mass_g", data=penguins)  # col: 넘긴 기준별로 그려준다.
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm",
            hue = "species", size = "body_mass_g",sizes=(30,100),
            data=penguins)  # size: 점의 크기를 시각적으로 구분해준다. sizes = (최소, 최대)
```

<h1> 라인플롯(Line Plot)</h1>

```python
flights = sns.load_dataset("flights")   # flights: 연도별 승객 수치
flights
```

```python
sns.relplot(x='year', y="passengers", kind='line', data=flights)  # kind: plot 종류 설정
```

```python
dots = sns.load_dataset('dots')   
dots
```

```python
sns.relplot(x="time",y="firing_rate", kind='line', data=dots)   
```

```python
# None 입력시 표시 삭제, "sd" 입력시 표준편차 
sns.relplot(x="time",y="firing_rate", kind='line', ci=None ,data=dots)                                                               
```

```python
# "sd" 입력시 표준편차 (standard deviation)
sns.relplot(x="time",y="firing_rate", kind='line', ci="sd" ,data=dots)
```

```python
# estimator=None: 추정회귀선 비활성화,
sns.relplot(x="time",y="firing_rate", estimator=None, 
            ci="sd", kind='line' ,data=dots) 
```

```python
# ci= "sd" 표준편차 활성화
sns.relplot(x="time",y="firing_rate", hue="choice", 
             ci="sd" ,kind='line',data=dots) 
```

```python
sns.relplot(x="time",y="firing_rate", hue="align",ci="sd" ,
            kind='line',data=dots) 
```

```python
sns.relplot(x="time",y="firing_rate", hue="align", 
             ci="sd", style='choice',
            kind='line', data=dots) 
```

```python
# dashes = False: 파선(- - -) 해제, markers = True: 기호가 들어간 점선 선택
sns.relplot(x="time",y="firing_rate", hue="align", 
            ci="sd", style='choice', 
            dashes=False, markers=True,
             kind='line',data=dots) 
```

```python
# col항목 열어 데이터 분리
sns.relplot(x="time",y="firing_rate", hue="align", 
            ci="sd", col='choice', 
            dashes=False, markers=True,
            kind='line', data=dots) 
```

```python
# query() 함수: pandas 에서 데이터프레임을 쿼리하고 원하는 조건에 맞는 데이터를 선택하는 데 사용되는 메서드이다. 
sns.relplot(x="time",y="firing_rate" ,
            style='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
# hue를 사용해서 데이터 세분화
sns.relplot(x="time",y="firing_rate" ,
            hue = "coherence",style='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
# col을 사용해서 데이터 세분화
sns.relplot(x="time",y="firing_rate" ,
            hue = "coherence",col='choice', 
            kind='line', data=dots.query("align == 'sacc'")) 
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x = "timepoint", y="signal",
          kind = 'line',data = fmri)
```

```python
# sort=False: 정렬 취소, seaborn이 데이터를 정렬시킨 후 보여준다.
sns.relplot(x = "timepoint", y="signal", sort=False,  # 
          kind = 'line',data = fmri)
```

```python
sns.relplot(x = "timepoint", y="signal",
            style='region',size='event',
          kind = 'line',data = fmri)
```

```python
# 많은 데이터를 같은 화면에 담아두는것은 좋지 않다는 예
sns.relplot(x = "timepoint", y="signal",
           hue="subject", style='region',size='event',
          kind = 'line',data = fmri)
```

```python
# col로 분리
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',size='event',
          kind = 'line',data = fmri)
```

```python
# palette: 색상 정의 후 적용
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',style='event',
           palette = palette, kind = 'line',data = fmri)
```

```python
# cue 데이터만 보기
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',
           palette = palette, kind = 'line',data = fmri.query("event == 'cue'"))
```

```python
# cue 데이터만 보기
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x = "timepoint", y="signal",
           hue="subject", col='region',row = "event",
           palette = palette, kind = 'line',data = fmri)
```
